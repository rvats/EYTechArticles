```python
# import required module
import os
import csv  

header = ['EDIFormName', 'PartnerSourceDataFile','PartnerDestinationDataFile', 'EDIVersion', 'SAPVersion', 'MapInformation_EDIFormName_EDIVersion_SAPVersion']

# assign directory
cwd = os.getcwd()

data = ['', '', '', '', '', '']

with open('EDIMapVersionDataInfo.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    
    # write the header
    writer.writerow(header)
    
    # Write Data Row
    # iterate over files in Current directory
    for mapDir in os.scandir(cwd):
        print(mapDir)
        if not mapDir.is_file():
            newDir = mapDir.path
            temp = mapDir.path[50:53]
            if temp[0] != '.':
                data[0] = temp
                for fileName in os.scandir(mapDir):
                    fName = os.path.basename(fileName.path)
                    data[1] = fName
                    data[2] = fName
                    with open(fileName.path, 'r') as file:
                            index = 0
                            try:
                                first_line = file.readline()
                                if '.dst' in fName:
                                    index = first_line.index('X*')
                                    edi = first_line[index+2:index+8]
                                    data[3]=edi
                                if '.src' in fName:
                                    index = first_line.index('SAP')
                                    sap = first_line[index:index+6]
                                    data[4]=sap
                            except:
                                print("An exception occurred while reading "+fName)
                    # write the data
                    print(data)
                    writer.writerow(data)
 
```
